```python
df["review_rating"] = df.groupby("category_product")["review_rating"].transform(lambda x:x.fillna(x.median()))
```

```python
frequency_mapping = {"Daily":1, "Weekly":7, "Biweekly":14, "Monthly": 30}
df["frequency_purchase"] = df ["frequency_of_purchase"].map(frequency_mapping)
df[["frequency_purchase","frequency_of_purchase"]].head(10)
```
